```python
import numpy as np
from plotnine import *
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import t
```

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-10, 10, 0.001)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,1,0.1))
plt.plot(x_axis, t.pdf(x_axis,1,0.1))
plt.show()
```

```python
t.pdf(10, 0.5,0.5)
```

```python
def zscore(x, m, s):
    return(min(max(0,(x-m)/s),1))
```

```python
zscore(20,0.5,0.3)
```

```python
def lineq(x):
    res = min(1, max(0,(2 * x - 1)))
    return(res)
```

```python
plt.plot(x_axis, [lineq(i) for i in x_axis])
plt.plot(x_axis, [zscore(i, 0.5,0.5) for i in x_axis])
```
